# CCM counter example

The following example shows that CCM may not be better than STD

```python
from graphviz import Digraph, Source
```

```python
dot = Digraph()

dot.node('X', 'X=[C, U1, U2, S]', style='filled', fillcolor='lightgray')
dot.node('C', 'C ~ N(0,1)', style='filled', fillcolor='lightgray')
dot.node('Y', 'Y=U1', style='filled', fillcolor='lightgray')
dot.node('U', 'U1=C+0.1*U2')
dot.node('Z', 'U2 ~ N(0, 1)')
dot.node('S', 'S=U2')
dot.edges(['UY', 'CU', 'ZU', 'UX', 'SX', 'CX', 'ZX'])
dot.edge('Z', 'S', style='dashed')

dot.graph_attr['rankdir'] = 'BT'
dot.render('figures/counter_example')
dot
```
