```python
def cub_shortcut_correlation(concepts, n_classes, show_concept_shortcut=False):
    # concepts: (200, n_concepts), n_classes: number of shortcut classes
    n_cub_classes, n_concepts = concepts.shape
    
    shortcuts = np.zeros((n_cub_classes, n_classes))
    for i in range(n_classes):
        shortcuts[range(i, n_cub_classes, n_classes), i] = 1
    if show_concept_shortcut:
        print(f'concepts (col 0:{n_concepts}), shortcuts (col {n_concepts}:{n_concepts+n_classes})')
        plt.matshow(np.hstack((concepts, shortcuts)))
        plt.show()
    
    a = np.hstack((np.array(concepts), shortcuts))
    cov_a = np.corrcoef(a.T)
    print(f'max abs correlation (c vs s): {np.abs(cov_a[n_concepts:, :n_concepts]).max():.2f}')

    plt.matshow(cov_a, cmap='seismic', vmax=1, vmin=-1)
    plt.show()
    
cub_shortcut_correlation(maj_concepts, 10, show_concept_shortcut=True)
```

```python
lowest_error_ind_maj, lowest_error = None, np.inf
for i in tqdm.trange(30):
    kmedoids = KMedoids(n_clusters=30, metric='correlation', init='k-medoids++').fit(maj_concepts.T)
    ind_maj_concepts = maj_concepts.iloc[:, kmedoids.medoid_indices_]
    abs_sum = np.array(np.abs(ind_maj_concepts.corr())).sum()
    if abs_sum < lowest_error:
        lowest_error = abs_sum
        lowest_error_ind_maj = ind_maj_concepts

ind_maj_concepts = lowest_error_ind_maj
ind_maj_concepts
```
